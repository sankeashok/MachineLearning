```
#importing required libraries
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading SampleSuperstore file and loadding to Dataframe

```
dataframe = pd.read_csv('/content/SampleSuperstore.csv',  header=0)
dataframe.head(10)  #displaying frst 10 rows of store data
```

```
#checking null values in the data set so that we can take action on the missing values etc.
dataframe.isnull().sum()
```

```
#Dataset glance
dataframe.info()
```

```
#checkngi f there are duplicate values in the dataset
dataframe.duplicated().sum()
dataframe.nunique()
```

```
# Visualising the State wise 
plt.figure(figsize=(15,15))
ed=sns.countplot(x="State", data=dataframe)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('State', fontsize=30, color = 'g',fontweight='bold') 
plt.ylabel('Count', fontsize=30, color = 'r',fontweight='bold')
plt.show()  
```

```
#Plot State vs Mean Sales
plt.figure(figsize=(30,18))
state = dataframe.groupby(['State'])['Sales'].mean().reset_index()
ed=sns.barplot(x = 'State', y='Sales', data=state)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('State', fontsize=20) 
plt.ylabel('Mean Sales', fontsize=20)
plt.show()
```

```
# Plot State vs Mean Profit
plt.figure(figsize=(20,10))
state1 = dataframe.groupby(['State'])['Profit'].mean().reset_index()
ed=sns.barplot(x = 'State', y='Profit', data=state1)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=10)
plt.xlabel('State', fontsize=30,color = 'y',fontweight='bold')
plt.ylabel('Mean Profit', fontsize=30,color = 'orange',fontweight='bold')
plt.show()
```

```
plt.figure(figsize=(10,5))
ed=sns.countplot(x="Category", data=dataframe)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Category', fontsize=20) 
plt.ylabel('Count', fontsize=20)
plt.show()  
```

```
#Correlation matrix
sns.heatmap(dataframe.corr(),annot=True)
plt.show()
```

```
# covariance matrix
sns.heatmap(dataframe.cov(),annot=True)
plt.show()
```

```
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit', data=dataframe , color='g',label='Discount')
plt.legend()
plt.show()
```

```
figsize=(15,10)
sns.pairplot(dataframe,hue='Sub-Category')
plt.show()
```

```
#Plot Sales vs Profit
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(dataframe["Sales"] , dataframe["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```
#Histogram
dataframe.hist(bins=50 ,figsize=(20,15))
plt.show();
```
